```python
import pandas as pd
import matplotlib.pyplot as plt
import os

from datetime import datetime as dt
def parse_time(ts):
    return dt.utcfromtimestamp(ts / 1000).strftime('%Y-%m-%d %H:%M:%S')

def get_last_file(seed = 0, period = "d", models = ["lr", "sgd", "rfr", "dummy", "arima"], reponse = "open"):
    # Daily: results/model_run_0_d_lr_rfr_sgd_arima_open_2021_08_15 22_14.json
    # Weekly: results/model_run_0_w_lr_rfr_sgd_arima_open_2021_08_15 20_39.json
    file_mask = f"model_run_{str(seed)}_{period}_{'_'.join(sorted(models))}_{reponse}"
    return sorted([f for f in os.listdir("./results/") if file_mask in f], reverse=True)[0]

def load_results(file=get_last_file()):
    print(f"Loading file: {file}")
    results = pd.read_json(f"results/{file}")
    results.columns = ["data"]
    results = pd.json_normalize(results["data"])
    results = results.set_index("id")
    return results

def load_all_results():
    full_df = pd.DataFrame()
    for f in os.listdir("results/"):
        if ".json" in f:
            full_df = pd.concat([full_df, load_results(f)])
    return full_df.reset_index()

full_results = load_all_results()
daily_results = load_results()
weekly_results = load_results(get_last_file(period="w"))

display(daily_results)
```

# Results selection

As we've performed multiple training cycles with identical data and random seeds, we noticed the results not to be identical. As such, we chose to filter results, selecting only those that do not have a 100% error rate.

```python
# number of samples for stratified sampling
N = 50
random_state = np.random.RandomState(seed=0)

def result_filter_rules(df, n=N, random_state=random_state):
    df = df[df["test_score"] < 1.0]
    df = df[df["model"] != "sgd"]
    df = df[~df.isna()]
    df = df[df["is_valid"] == True]
    df = df.groupby(["model"], group_keys=False).apply(lambda x: x.sample(n, random_state=random_state)).copy()
    return df


# comparison dims
covid_v_non_covid = ["full covid data", "weather data only"]
covid_v_infected = ["full covid data", "infected data only"]
covid_v_deceased = ["full covid data", "deceased data only"]
covid_v_vaccinated = ["full covid data", "vaccinated data only"]
covid_v_stringency = ["full covid data", "stringency data only"]
covid_v_index = ["full covid data", "index data only"]



# Daily results
base_daily_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_12 12_05.json"))

## covid
daily_covid_results = base_daily_results.copy()
daily_covid_results["data_type"] = covid_v_non_covid[0]

## weather only (non-covid)
daily_non_covid_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_05 13_42.json"))
daily_non_covid_results["data_type"] = covid_v_non_covid[1]

## infected
daily_infected_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_05 16_29.json"))
daily_infected_results["data_type"] = covid_v_infected[1]

## deceased
daily_deceased_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_11 15_15.json"))
daily_deceased_results["data_type"] = covid_v_deceased[1]

## vaccinated
daily_vaccinated_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_11 16_57.json"))
daily_vaccinated_results["data_type"] = covid_v_vaccinated[1]

## stringency
daily_stringency_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_17 17_36.json"))
daily_stringency_results["data_type"] = covid_v_stringency[1]

## index only
daily_index_results = result_filter_rules(load_results("model_run_0_d_arima_dummy_lr_rfr_sgd_open_2021_12_27 15_14.json"))
daily_index_results["data_type"] = covid_v_index[1]

# Weekly results
base_weekly_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_04 14_19.json"))

## covid
weekly_covid_results = base_weekly_results.copy()
weekly_covid_results["data_type"] = covid_v_non_covid[0]

## weather only (non-covid)
weekly_non_covid_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_05 14_36.json"))
weekly_non_covid_results["data_type"] = covid_v_non_covid[1]

## infected
weekly_infected_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_05 16_46.json"))
weekly_infected_results["data_type"] = covid_v_infected[1]

## deceased
weekly_deceased_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_11 15_31.json"))
weekly_deceased_results["data_type"] = covid_v_deceased[1]

## vaccinated
weekly_vaccinated_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_11 17_31.json"))
weekly_vaccinated_results["data_type"] = covid_v_vaccinated[1]

## stringency
weekly_stringency_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_17 17_53.json"))
weekly_stringency_results["data_type"] = covid_v_stringency[1]

## index only
weekly_index_results = result_filter_rules(load_results("model_run_0_w_arima_dummy_lr_rfr_sgd_open_2021_12_27 15_28.json"))
weekly_index_results["data_type"] = covid_v_index[1]
```

```python
from scipy.stats import ttest_ind

alternative="two-sided"

def perform_mean_hypothesis_test(df, dims, col="data_type", random_state=random_state, alternative=alternative):
    scores = [df[df[col] == d].test_score.mean() for d in dims]
    a = df[df[col] == dims[0]].test_score
    b = df[df[col] == dims[1]].test_score
    
    print(f"The mean test scores for {dims[0]} vs. {dims[1]} are:")
    for i, d in enumerate(dims):
        print(f"{d} mean test score: {scores[i]}")
    
    print(f"The T-test score for the difference in means is:")
    print(ttest_ind(a, b, random_state=random_state, alternative=alternative))
```
